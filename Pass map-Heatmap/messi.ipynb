```python
# read data 
df = pd.read_csv('G:/football/passmap-main/messibetis.csv')
df
```

```python
# resize x,y data

df['x'] = df['x'] * 1.2
df['endX'] = df['endX'] * 1.2
df['y'] = df['y'] * 0.8
df['endY'] = df['endY'] * 0.8
```

```python
df.head()
```

```python
#scatter plot for x,y points
pitch = Pitch(pitch_type='statsbomb', pitch_color= 'gray',line_color='white',
              stripe=False, shade_middle=False, shade_color='blue', positional = False, axis = True, label = False)
pitch.draw(figsize = (13.5, 7))
plt.scatter(df['x'], df['y'], color = 'yellow')
pitch.draw(figsize = (13.5, 7))
plt.scatter (df['endX'], df['endY'], color = 'navy');
```

```python
# for loop to change points color to yellow / red
pitch.draw(figsize = (13.5, 7))
for x in range(len(df['x'])):
    
    if df['outcome'][x] == 'Successful':
        plt.scatter( df['x'][x], df['y'][x], color = 'yellow')
    if df['outcome'] [x]== 'Unsuccessful':
        plt.scatter( df['x'][x], df['y'][x], color = 'red')
    
```

```python
# Line plot passes from start to end
pitch.draw(figsize = (13.5, 7))
for x in range(len(df['x'])):
    plt.plot((df['x'][x], df['endX'][x]),(df['y'][x] , df['endY'][x]), color = 'green')
```

```python
# plot passes using for loop and green\red color
#fig, axs = pitch.draw(figsize=(18,12),nrows=1, ncols=3)

#for ax in axs.flat:
#plt.figure()
#plt.subplot()
pitch.draw(figsize=(12,6))
plt.title('Messi Successful passes aganist Real Beties')
plt.savefig('Messi Successful passes aganist Real Beties.png')

for x in range(len(df['x'])):
    
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x], df['endX'][x]),(df['y'][x] , df['endY'][x]), color = 'yellow')
        plt.scatter( df['x'][x], df['y'][x], color = 'yellow')

#plt.subplot()
pitch.draw(figsize=(12,6))
plt.title('Messi Unsuccessful passes aganist Real Beties')
for x in range(len(df['x'])):
    if df['outcome'] [x]== 'Unsuccessful':
        plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color = 'red')
        plt.scatter( df['x'][x], df['y'][x], color = 'red')
plt.savefig('Messi Unsuccessful passes aganist Real Beties.png')        
#plt.subplot()        
pitch.draw(figsize=(12,6))
plt.title('Messi passes aganist Real Beties')

for x in range(len(df['x'])):
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x], df['endX'][x]),(df['y'][x] , df['endY'][x]), color = 'yellow')
        plt.scatter( df['x'][x], df['y'][x], color = 'yellow')

    if df['outcome'] [x]== 'Unsuccessful':
        plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color = 'red')
        plt.scatter( df['x'][x], df['y'][x], color = 'red')
plt.savefig('Messi full passes aganist Real Beties.png');
```

```python
# find number of unsuccessful passes
unsucc = (df['outcome'] == 'Unsuccessful')
unsucc.sum()
```

```python
# successful passes percentage
correct_pass_perc = ((len(df)- unsucc.sum()) / len(df)) * 100
print (f'successful passing percentage = {correct_pass_perc : .2f}%')
```

```python
# unsuccessful passes percentage
worng_pass_perc = (unsucc.sum() / len(df)) *100
print (f'Unsuccessful passing percentage = {worng_pass_perc : .2f}%')
```
