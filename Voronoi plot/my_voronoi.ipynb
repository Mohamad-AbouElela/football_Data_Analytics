```python
import pandas as pd
import numpy as np
from mplsoccer import VerticalPitch
import matplotlib.pyplot as plt

%matplotlib inline
%config Completer.use_jedi = False
```

```python
# read file
df = pd.read_csv('voronoi.csv')
df
```

```python
# prepare figure 
fig, ax = plt.subplots(figsize = (13, 8))
fig.set_facecolor('#224554')
ax.set_facecolor('#224554')

pitch = VerticalPitch(pitch_color='#224554', half=True, tight_layout=True)
pitch.draw(ax = ax)
```

```python
# create points for voronoi plot
points = np.column_stack((df.x, df.y))
points
```

```python
fig, ax = plt.subplots(figsize = (13, 8))
fig.set_facecolor('#224554')
ax.set_facecolor('#224554')

pitch = VerticalPitch(pitch_color='#224554', half=True, tight_layout=True)
pitch.draw(ax = ax)

x = df.x
y = df.y
team1, team2 = pitch.voronoi(x, y ,df.team)
t1 = pitch.polygon(team1, ax= ax, fc = 'red', ec = 'white', alpha = 0.6)
t2 = pitch.polygon(team2, ax= ax, fc = 'blue', ec= 'black', alpha = 0.6)

for i in range(len(df.x)):
    if df.team[i] == 0:
        pitch.scatter(df.x[i],df.y[i], color = 'black', s= 75, ax = ax)
    if df.team[i] == 1:
        pitch.scatter(df.x[i], df.y[i], color = 'white', ax = ax)
plt.title('Cadez VS Barcelona voronoi plot', color = 'white', fontsize = 18)
plt.savefig('my_voroni.png');
```
