```python
# import required libraries

import pandas as pd 
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch,VerticalPitch

%matplotlib inline
```

```python
# read file (Barca shots VS Juvy)
df = pd.read_csv('shotmaps.csv')
df
```

```python
fig, ax = plt.subplots(figsize = (12, 8))
pitch = VerticalPitch(half = True,
                      pitch_color = 'grass', 
                    axis = False, label = False,
                     stripe = False, stripe_color = '#73D629')



pitch.draw(ax=ax)
plt.gca().invert_yaxis()

for x in range(len(df)):
    if df['outcome'][x] == 'Missed':
        plt.scatter(df['x'][x], df['y'][x], s = 100, c = 'red', alpha = 0.9)
        
    if df['outcome'][x] == 'Saved':
        plt.scatter(df['x'][x], df['y'][x],  s = 100, c ='blue', alpha = 0.9)
     
    if df['outcome'][x] == 'Post':
        plt.scatter(df['x'][x], df['y'][x],  s = 100, c = 'yellow', alpha = 0.9)
        
plt.title('Barca Shots VS Juvy', fontsize = 22)
plt.savefig('barca shots.jpeg');
```
