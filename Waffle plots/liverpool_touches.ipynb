```python
# Import required libiraries

import pandas as pd
import matplotlib.pyplot as plt
from pywaffle.waffle import Waffle
from highlight_text import fig_text

%matplotlib inline
%config Completer.use_jedi = False
```

```python
# read file

df = pd.read_csv('liv_touches.csv')
df
```

```python
# columns names

my_list = df.columns.values.tolist()
```

```python
# drop unwanted columns

df.drop(columns=['Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10',
       'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14',
       'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17'], axis = 1, inplace = True)
```

```python
df.drop('Live', axis = 1, inplace = True)
df
```

```python
# split players names

df['Player'] = df['Player'].str.split('\\',expand =True)[0]
df
```

```python
# Arrange df descendeing by Touches

df.sort_values(by = ['Touches'], ascending = False, inplace=True)
df.drop(14, inplace=True)
df
```

```python
# remove touches column

df.drop('Touches', axis = 1, inplace = True)
```

```python
my_list = df.columns.values.tolist()
my_list
```

```python
# Transpose df

df_t = df.transpose()
df_t.columns = df_t.iloc[0]
```

```python
df_t.drop('Player',axis = 0, inplace=True)
```

```python
df_t
```

```python
my_list_t = df_t.columns.values.tolist()
my_list_t
```

```python
df_new = pd.DataFrame(df_t, columns=my_list_t)
```
